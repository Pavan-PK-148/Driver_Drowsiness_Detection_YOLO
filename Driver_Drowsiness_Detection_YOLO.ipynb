```python
import os
import shutil
import random

train_path = "mrl_dataset/data/train"
val_path = "mrl_dataset/data/val"

# Class mapping
classes = {
    "awake": 0,    # open_eye
    "sleepy": 1    # closed_eye
}

def convert_split(src_path, split_type):
    for class_name in classes:
        class_id = classes[class_name]
        class_folder = os.path.join(src_path, class_name)

        images = os.listdir(class_folder)

        for img in images:
            src_img = os.path.join(class_folder, img)
            dst_img = os.path.join("dataset/images", split_type, img)
            dst_label = os.path.join("dataset/labels", split_type, img.rsplit(".",1)[0] + ".txt")

            # Copy image
            shutil.copy(src_img, dst_img)

            # Create YOLO label file
            with open(dst_label, "w") as f:
                f.write(f"{class_id} 0.5 0.5 1.0 1.0\n")

# Convert train and val
convert_split(train_path, "train")
convert_split(val_path, "val")

print("Conversion Complete âœ…")
```

```python
import os
import shutil

# Base paths
base_path = "/content/drive/MyDrive/drowsiness_project"
train_path = os.path.join(base_path, "data/train")
val_path = os.path.join(base_path, "data/val")
yolo_path = os.path.join(base_path, "yolo_dataset")

# Class mapping
classes = {
    "awake": 0,
    "sleepy": 1
}

def convert_split(src_path, split):
    for class_name, class_id in classes.items():
        class_folder = os.path.join(src_path, class_name)
        images = os.listdir(class_folder)

        for img in images:
            src_img = os.path.join(class_folder, img)
            dst_img = os.path.join(yolo_path, "images", split, img)
            dst_label = os.path.join(yolo_path, "labels", split, img.rsplit(".",1)[0] + ".txt")

            # Copy image
            shutil.copy(src_img, dst_img)

            # Create YOLO label file (full image bounding box)
            with open(dst_label, "w") as f:
                f.write(f"{class_id} 0.5 0.5 1.0 1.0\n")

# Convert train and val
convert_split(train_path, "train")
convert_split(val_path, "val")

print("YOLO dataset conversion complete âœ…")
```
